```python
# TODO 08 - what is the proportion of total sales (%) in West + Central in 2019 e.g. 25% 
```

```python
# The data of sales in 2019
t = df[pd.to_datetime(df['Order Date']).dt.year == 2019]
t
```

```python
# the proportion of each region
pts = t.groupby('Region')["Sales"].sum()/t['Sales'].sum()*100
pts.reset_index()
```

```python
# the total sales (%) in West + Central
pts.iloc[[0,3]].sum()
```

**The proportion of total sales (%) in West + Central in 2019 is 54.98%**

```python
# TODO 09 - find top 10 popular products in terms of number of orders vs. total sales during 2019-2020
```

```python
# the data sales in 2019-2020
pop = df[(pd.to_datetime(df['Order Date']).dt.year == 2019) | (pd.to_datetime(df['Order Date']).dt.year == 2020)]
pop
```

```python
# 10 popular product in terms of number of orders
pop.groupby('Product Name')["Order ID"].count().reset_index().sort_values('Order ID',ascending=False).head(10)
```

```python
# 10 popular product in terms of total sales
pop.groupby('Product Name')["Sales"].sum().reset_index().sort_values('Sales',ascending=False).head(10)
```

```python
# TODO 10 - plot at least 2 plots, any plot you think interesting :)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
s = pop.groupby('Product Name')["Order ID"].count().reset_index().sort_values('Order ID',ascending=False).head(10)

fig = plt.gcf()
fig.set_size_inches(10, 6)
plt.grid(color="black", linewidth=0.5)
sns.barplot(y='Product Name',x='Order ID',data=s , orient='h');
```

```python
fig = plt.gcf()
fig.set_size_inches(10, 10)
plt.pie(x=s['Order ID'], labels=s['Product Name'],autopct='%.2f%%');
```
